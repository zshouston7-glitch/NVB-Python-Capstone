## 📌 Project Overview
This project is an interactive command-line **Recipe Manager** (Option B3). It addresses the problem of tracking household recipes, modifying portion quantities effortlessly, and searching for inspiration based on ingredients already available in the kitchen.

### Key Features Demonstrated:
* **Nested Dictionaries:** Master recipe catalog storage mapping keys to multi-valued records.
* **Indefinite Control Loops:** User-driven command interface that safely persists session states.
* **Modular Architecture:** Custom standalone functions complete with descriptive `docstrings`.

```python
"""
NVB Python Summer 2026 Capstone Project
Project: Interactive Recipe Manager (Option B3)
Author: Zoey Houston
"""

# Initial data structure: Pre-loaded with Chocolate Chip Cookies, Blueberry Muffins, and Blackberry Cobbler
recipe_book = {
    "Chocolate Chip Cookies": {
        "ingredients": ["1 cup butter", "1 cup sugar", "2 cups flour", "1 cup chocolate chips", "2 eggs"],
        "steps": ["Cream butter and sugar together.", "Beat in the eggs.", "Mix in the flour and fold in chocolate chips.", "Bake at 350°F (175°C) for 10 minutes."],
        "servings": 24
    },
    "Blueberry Muffins": {
        "ingredients": ["1.5 cups flour", "0.75 cup sugar", "0.5 cup milk", "0.33 cup vegetable oil", "1 cup fresh blueberries", "1 egg"],
        "steps": ["Combine dry ingredients in a large bowl.", "Whisk milk, oil, and egg together; stir into dry ingredients.", "Gently fold in blueberries.", "Divide into muffin tin and bake at 400°F (200°C) for 20 minutes."],
        "servings": 12
    },
    "Blackberry Cobbler": {
        "ingredients": ["4 cups blackberries", "1 cup sugar", "1 cup flour", "1 cup milk", "0.5 cup melted butter"],
        "steps": ["Rinse blackberries and toss with 1/4 cup sugar; place in baking dish.", "Whisk flour, remaining sugar, milk, and melted butter into a smooth batter.", "Pour batter evenly over the blackberries.", "Bake at 375°F (190°C) for 45 minutes until golden brown."],
        "servings": 8
    }
}

def add_recipe(name, ingredients, steps, servings):
    """
    Adds a new recipe to the global recipe book.

    Args:
        name (str): The name of the recipe.
        ingredients (list): A list of ingredient strings.
        steps (list): A list of instruction strings.
        servings (int): The default number of servings.
    """
    recipe_book[name.title()] = {
        "ingredients": ingredients,
        "steps": steps,
        "servings": int(servings)
    }
    print(f"\n✅ Success: '{name.title()}' has been added to your Recipe Manager!")

def display_recipe(name, scale_servings=None):
    """
    Prints a formatted, easy-to-read version of a specific recipe.
    Can optionally scale ingredients based on a target serving size.

    Args:
        name (str): The name of the recipe to display.
        scale_servings (int, optional): The new target serving count.
    """
    recipe_name = name.title()
    if recipe_name not in recipe_book:
        print("\n❌ Error: Recipe not found.")
        return

    recipe = recipe_book[recipe_name]
    current_servings = recipe["servings"]

    print("\n" + "="*40)
    print(f"📖 RECIPE: {recipe_name.upper()}")
    print("="*40)

    if scale_servings and scale_servings != current_servings:
        scale_factor = scale_servings / current_servings
        print(f"👥 Servings: {scale_servings} (Scaled from {current_servings}x)")
        print("\n🛒 Ingredients:")
        for ing in recipe["ingredients"]:
            print(f"  • {ing} (Multiply amounts by {scale_factor:.1f}x)")
    else:
        print(f"👥 Servings: {current_servings}")
        print("\n🛒 Ingredients:")
        for ing in recipe["ingredients"]:
            print(f"  • {ing}")

    print("\n👩‍🍳 Steps:")
    for idx, step in enumerate(recipe["steps"], 1):
        print(f"  {idx}. {step}")
    print("="*40 + "\n")

def search_by_ingredient(ingredient):
    """
    Searches all recipes and prints a list of recipes containing a specific ingredient.

    Args:
        ingredient (str): The partial or full name of an ingredient to find.
    """
    search_term = ingredient.lower()
    found_recipes = []

    for name, details in recipe_book.items():
        for ing in details["ingredients"]:
            if search_term in ing.lower():
                found_recipes.append(name)
                break

    print("\n" + "-"*40)
    if found_recipes:
        print(f"🔍 Recipes containing '{ingredient}':")
        for recipe in found_recipes:
            print(f"  ⭐ {recipe}")
    else:
        print(f"❌ No recipes found containing '{ingredient}'.")
    print("-"*40 + "\n")

def main():
    """
    The main control loop for the interactive command-line interface.
    """
    print("✨ Welcome to your Capstone Recipe Manager ✨")

    while True:
        print("Main Menu:")
        print("1. ➕ Add a New Recipe")
        print("2. 🔍 Search Recipes by Ingredient")
        print("3. 📖 View and Scale a Recipe")
        print("4. 📋 List All Saved Recipes")
        print("5. ❌ Exit")

        choice = input("\nChoose an option (1-5): ").strip()

        if choice == "1":
            name = input("Enter recipe name: ").strip()
            ing_input = input("Enter ingredients separated by commas: ")
            ingredients = [i.strip() for i in ing_input.split(",") if i.strip()]
            steps_input = input("Enter steps separated by periods: ")
            steps = [s.strip() for s in steps_input.replace(".", ";").split(";") if s.strip()]
            servings = input("How many servings does this make? ")
            servings = int(servings) if servings.isdigit() else 1

            add_recipe(name, ingredients, steps, servings)

        elif choice == "2":
            ing_search = input("Enter an ingredient to search for (e.g. blackberry): ").strip()
            search_by_ingredient(ing_search)

        elif choice == "3":
            name = input("Enter the recipe name to view: ").strip()
            scale_choice = input("Do you want to scale the servings? (yes/no): ").lower().strip()

            if scale_choice in ["yes", "y"]:
                new_servings = input("Enter new number of servings: ")
                new_servings = int(new_servings) if new_servings.isdigit() else None
                display_recipe(name, scale_servings=new_servings)
            else:
                display_recipe(name)

        elif choice == "4":
            print("\n📚 Current Recipe Book:")
            for r in recipe_book:
                print(f"  • {r}")
            print()

        elif choice == "5":
            print("\n👋 Happy Cooking! Goodbye.")
            break
        else:
            print("\n⚠️ Invalid selection. Please choose 1-5.\n")

if __name__ == "__main__":
    main()
```
